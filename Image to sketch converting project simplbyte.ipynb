```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2
```

```python
image=cv2.imread('image11.jpeg')
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(8,8))
plt.imshow(image)
plt.title("original image")
plt.show()
```

```python
Gray_image =cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(8,8))
plt.imshow(Gray_image,cmap='gray')
plt.title('Gray_image')
plt.axis("off")
plt.show()
```

```python
inverted_image = cv2.bitwise_not(Gray_image)
plt.figure(figsize=(8,8))
plt.imshow(inverted_image,cmap='gray')
plt.title("inverted image")
plt.axis("off")
plt.show()
```

```python
smoothening_image = cv2.GaussianBlur(inverted_image,(21,21),sigmaX=0,sigmaY=0)
plt.figure(figsize=(8,8))
plt.imshow(smoothening_image,cmap='gray')
plt.title("smoothening image")
plt.axis("off")
plt.show()
```

```python
pencil_sketch = cv2.divide(Gray_image,255-smoothening_image,scale=255)
plt.figure(figsize=(8,8))
plt.imshow(pencil_sketch,cmap='gray')
plt.title("pencil sketch")
plt.axis("off")
plt.show()
```
